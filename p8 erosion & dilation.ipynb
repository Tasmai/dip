```python
#erosion
import cv2 as cv
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import skimage.exposure

# load image
img = cv.imread('download.jpg')

def bgr2rgb(img):
    return cv.cvtColor(img,cv.COLOR_BGR2RGB)

def plot_image(img1, img2, img3, img4):
    
    f, axarr = plt.subplots(2,2,figsize = (15,10))
    
    axarr[0,0].imshow(bgr2rgb(img1))
    axarr[0,1].imshow(bgr2rgb(img2))
    axarr[1,0].imshow(bgr2rgb(img3))
    axarr[1,1].imshow(bgr2rgb(img4))
    
    
#convert to grayscale
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

#threshold to get a binary image
img_tresh = cv.threshold(img_gray, 120, 255, cv.THRESH_BINARY)[1]

# make 2 kernel sizes
kernel = np.ones((5, 5), np.uint8)
kernel2 = np.ones((11, 11), np.uint8)

#eroding
img_erod = cv.erode(img_tresh,kernel,iterations=1)
img_erod2 = cv.erode(img_tresh,kernel2,iterations=1)


plot_image(img_gray,img_tresh,img_erod,img_erod2)
```

```python
# dilation

# load image
img = cv.imread('download.jpg')

def bgr2rgb(img):
    return cv.cvtColor(img,cv.COLOR_BGR2RGB)

def plot_image(img1, img2, img3, img4):
    
    f, axarr = plt.subplots(2,2,figsize = (15,10))
    
    axarr[0,0].imshow(bgr2rgb(img1))
    axarr[0,1].imshow(bgr2rgb(img2))
    axarr[1,0].imshow(bgr2rgb(img3))
    axarr[1,1].imshow(bgr2rgb(img4))

#convert to grayscale
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
#threshold to get a binary image
img_tresh = cv.threshold(img_gray, 120, 255, cv.THRESH_BINARY)[1]

# make 2 kernel sizes
kernel = np.ones((5, 5), np.uint8)
kernel2 = np.ones((11, 11), np.uint8)

#eroding
img_dilate = cv.dilate(img_tresh,kernel,iterations=1)
img_dilate2 = cv.dilate(img_tresh,kernel2,iterations=1)


plot_image(img_gray,img_tresh,img_dilate,img_dilate)
```
