```python
import cv2 as cv
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import skimage.exposure

#importing the image
img = cv.imread("download.jpg")

def bgr2rgb(img):
    return cv.cvtColor(img,cv.COLOR_BGR2RGB)

#converting to gray scale
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)


#Performing laplacian 
lap = cv.Laplacian(gray,cv.CV_64F,ksize = 1)
lap = np.absolute(lap)
lap = np.array(lap, dtype = "uint8")

#Plot image
fx,ax = plt.subplots(2,figsize = (15,10))

ax[0].imshow(bgr2rgb(gray))
ax[1].imshow(bgr2rgb(lap))
```

```python
import cv2 as cv
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import skimage.exposure

#reading image
img = cv.imread('download.jpg')

def bgr2rgb(img):
    return cv.cvtColor(img,cv.COLOR_BGR2RGB)

#blurring the image
blur = cv.blur(img, ksize = (7,7))

#turning it into grayscale
gray = cv.cvtColor(blur, cv.COLOR_BGR2GRAY)

def plot_image(img1, img2, img3, img4):
    
    f, axarr = plt.subplots(2,2,figsize = (15,10))
    
    axarr[0,0].imshow(bgr2rgb(img1))
    axarr[0,1].imshow(bgr2rgb(img2))
    axarr[1,0].imshow(bgr2rgb(img3))
    axarr[1,1].imshow(bgr2rgb(img4))

#Performing laplacian on blurred image
lap = cv.Laplacian(gray, cv.CV_64F,ksize = 3)
lap = np.uint8(np.absolute(lap))

#Sharpening the original image
new_img = np.clip(lap+gray,0,np.max(gray))

plot_image(blur,gray,lap,new_img)
```
