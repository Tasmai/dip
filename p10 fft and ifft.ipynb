```python
import cv2 as cv
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import skimage.exposure

# Read the image in grayscale
img = cv.imread('download.jpg', 0)

def bgr2rgb(img):
    return cv.cvtColor(img,cv.COLOR_BGR2RGB)

# Calculate the 2D FFT of the image
f = np.fft.fft2(img)

# Shift the zero-frequency component to the center of the spectrum
fshift = np.fft.fftshift(f)

# Calculate the magnitude spectrum
magnitude_spectrum = 20*np.log(np.abs(fshift))

# Display the original and magnitude spectrum images
dx,ax = plt.subplots(2,figsize = (15,10))
ax[0].imshow(bgr2rgb(img))
ax[1].imshow(magnitude_spectrum, cmap='gray')
```
