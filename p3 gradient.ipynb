```python
img = cv.imread('download.jpg', cv.IMREAD_GRAYSCALE)
Kx = np.array([[-1, 0, 1], 
               [-2, 0, 2], 
               [-1, 0, 1]])
Ky = np.array([[1,   2,  1], 
               [0,   0,  0], 
              [-1,  -2, -1]])

grad_x = cv.filter2D(img, -1, Kx)
grad_y = cv.filter2D(img, -1, Ky)

magnitude = np.sqrt(grad_x*2 + grad_y*2)
grad_x[grad_x == 0] = 1
angle = np.arctan(grad_y/grad_x)

fx,ax = plt.subplots(4, figsize = (30,15))

ax[0].imshow(grad_x,cmap = 'gray')
ax[1].imshow(grad_y,cmap = 'gray')
ax[3].imshow(angle,cmap = 'gray')
ax[2].imshow(np.uint8(magnitude),cmap = 'gray')
```
