```python
import cv2 as cv
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import skimage.exposure

#using previus images
img = cv.imread("download.jpg")
#creating a filter 
size = int(input("Enter the size of the kernel: "))

def bgr2rgb(img):
    return cv.cvtColor(img,cv.COLOR_BGR2RGB)

def plot_image(img1, img2, img3, img4):
    
    f, axarr = plt.subplots(2,2,figsize = (15,10))
    
    axarr[0,0].imshow(bgr2rgb(img1))
    axarr[0,1].imshow(bgr2rgb(img2))
    axarr[1,0].imshow(bgr2rgb(img3))
    axarr[1,1].imshow(bgr2rgb(img4))

#averaging using the filter
blur = cv.blur(img,ksize = (size,size))
box = cv.boxFilter(img,-1,(size,size))
plot_image(img,img,box,blur)
```
